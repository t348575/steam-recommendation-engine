```python
data = pd.read_csv('popular_tags_freq.csv')["0"].to_list()
temp = Counter(data).most_common(20)
twenty_most_common = set()
for i in temp:
    twenty_most_common.add(i[0])
data = list(filter(lambda x: x in twenty_most_common, data))
fig, axs = plt.subplots(1, 1,figsize =(20, 20),tight_layout = True)
axs.grid(b = True, color ='grey',linestyle ='-.', linewidth = 0.5,alpha = 0.6)
axs.bar(range(20), 20, width=1, align='center')
axs.hist(data, histtype='bar', ec='black', bins=20)
```
