# Code - Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
# pd.set_option('display.max_colwidth', None)
```

## Word Cloud of Popluar Tags

```python
fig = plt.figure(figsize=(20, 20))
df = pd.read_csv('./orignal_steam_games.csv')

AllTags = []
comments_words=""
Tags = df["popular_tags"].tolist()
for i in range(len(Tags)):
    t = Tags[1].split(",")
    AllTags.extend(t)
    comments_words += " ".join(t)+" "

wordcloud = WordCloud(width = 3000, height = 2000,
                background_color ='white',
                min_font_size = 10).generate(comments_words)
plt.axis("off")
plt.imshow(wordcloud);
```

## A Pie Chart to visualize the Pricing of Games

```python
df = pd.read_csv('./steam_games_p1.csv')
df[df['original_price'].isna()] = "Free"
Price = df['original_price'].tolist()
for i in range(len(Price)):
    if '$' not in Price[i]:
        Price[i] = 'FREE'
    else:
        Price[i] = Price[i][1:]
for i in range(len(Price)):
    if(Price[i] != "FREE"):
        Price[i] = float(Price[i])
    else:
        Price[i] = 0

for i in range(len(Price)):
    if(Price[i] <= 10):
        Price[i] = 10
    elif (Price[i] > 10 and Price[i] <= 20):
        Price[i] = 20
    elif (Price[i] > 20 and Price[i] <= 30):
        Price[i] = 30
    elif (Price[i] > 30 and Price[i] <= 40):
        Price[i] = 40
    elif (Price[i] > 40 and Price[i] <= 50):
        Price[i] = 50
    elif (Price[i] > 50 and Price[i] <= 60):
        Price[i] = 60
    elif (Price[i] > 60 and Price[i] <= 70):
        Price[i] = 70
    elif (Price[i] > 70 and Price[i] <= 80):
        Price[i] = 80
    elif (Price[i] > 80 and Price[i] <= 90):
        Price[i] = 90
    elif (Price[i] > 90 and Price[i] <= 100):
        Price[i] = 100
    elif (Price[i] > 100 and Price[i] <= 150):
        Price[i] = 150
    elif (Price[i] > 150 and Price[i] <= 200):
        Price[i] = 200
    elif (Price[i] > 250 and Price[i] <= 300):
        Price[i] = 300
    elif (Price[i] > 300 and Price[i] <= 350):
        Price[i] = 350
    elif (Price[i] > 350 and Price[i] <= 400):
        Price[i] = 400
    elif (Price[i] > 400 and Price[i] <= 450):
        Price[i] = 450
    elif (Price[i] > 500 and Price[i] <= 550):
        Price[i] = 550
    elif (Price[i] > 550 and Price[i] <= 600):
        Price[i] = 600
    elif (Price[i] > 600 and Price[i] <= 650):
        Price[i] = 650
    elif (Price[i] > 650 and Price[i] <= 700):
        Price[i] = 700
    elif (Price[i] > 700 and Price[i] <= 750):
        Price[i] = 750
    elif (Price[i] > 750 and Price[i] <= 800):
        Price[i] = 800
    elif (Price[i] > 800 and Price[i] <= 850):
        Price[i] = 850
    elif (Price[i] > 850 and Price[i] <= 900):
        Price[i] = 900
    elif (Price[i] > 900 and Price[i] <= 950):
        Price[i] = 950
    elif (Price[i] > 950 and Price[i] <= 1000):
        Price[i] = 1000
    else:
        Price[i] = 0

PriceSet = list(set(Price))
Count = []
for i in PriceSet:
    Count.append(Price.count(i))


fig = plt.figure(figsize =(10, 80))
patches, texts = plt.pie(Count, startangle=90, radius=1.2)
plt.legend(patches, PriceSet, loc='center', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
plt.title('Price Split')
plt.show()
```
